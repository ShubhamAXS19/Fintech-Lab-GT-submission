```python
# Renaming

import os

# Define the path to the 10-K directory
directory_path = '/content/sec-edgar-filings/MSFT/10-K'

# Loop through each folder in the directory
for folder_name in os.listdir(directory_path):
    # Split the folder name at '-'
    parts = folder_name.split('-')
    if len(parts) >= 2:
        # Extract the year from the folder name
        year = parts[1][:2]  # Assuming the year is in the second part, adjust as needed
        # Construct the new folder name
        new_folder_name = f"{year}"
        # Construct the full paths
        old_path = os.path.join(directory_path, folder_name)
        new_path = os.path.join(directory_path, new_folder_name)
        # Rename the folder
        os.rename(old_path, new_path)
        print(f"Renamed '{folder_name}' to '{new_folder_name}'")
    else:
        print(f"Skipping folder '{folder_name}' as it does not match the expected format")
```

```python
# Converting all sec filling to .txt file

import os
from bs4 import BeautifulSoup
import re

def convert_html_to_text(html_content):
    soup = BeautifulSoup(html_content, 'html.parser')
    plain_text = soup.get_text()
    plain_text = re.sub(r'\s+', ' ', plain_text).strip()
    return plain_text

def convert_files_in_directory(directory):
    for root, _, files in os.walk(directory):
        for file in files:
            if file.endswith('.txt'):
                file_path = os.path.join(root, file)
                with open(file_path, 'r', encoding='utf-8') as f:
                    html_content = f.read()

                plain_text_content = convert_html_to_text(html_content)

                # Rename the .txt file to _converted.txt
                filename, _ = os.path.splitext(file)
                new_file_name = filename + '_converted.txt'
                new_file_path = os.path.join(root, new_file_name)

                # Write the converted plain text to the new file
                with open(new_file_path, 'w', encoding='utf-8') as f:
                    f.write(plain_text_content)

                print(f"Conversion complete. Converted text saved to {new_file_path}")

# Replace 'path_to_10k_directory' with the path to your '10-k' directory
directory_path = '/content/sec-edgar-filings/MSFT/10-K'
convert_files_in_directory(directory_path)
```

```python
from fpdf import FPDF
import os

def convert_html_to_text(html_file_path, output_dir, year):
    # Read HTML content from the file
    with open(html_file_path, 'r', encoding='utf-8') as html_file:
        html_content = html_file.read()

    # Parse HTML content and prettify to format it
    soup = BeautifulSoup(html_content, 'lxml')
    prettified_html = soup.prettify()

    # Extract text from prettified HTML
    plain_text = BeautifulSoup(prettified_html, 'html.parser').get_text()

    # Get the filename without extension
    filename = os.path.splitext(os.path.basename(html_file_path))[0]

    # Save the converted text to a new text file in the output directory
    output_text_file_path = os.path.join(output_dir, f"{filename}-{year}.txt")
    with open(output_text_file_path, 'w', encoding='utf-8', errors='ignore') as output_file:
        output_file.write(plain_text)

    print(f"Text conversion complete. Converted text saved to {output_text_file_path}")

    # Convert text file to PDF
    output_pdf_file_path = os.path.join(output_dir, f"{filename}-{year}.pdf")
    pdf = FPDF()
    pdf.add_page()
    pdf.set_font("Arial", size=12)

    # Convert text to Unicode escape sequences
    with open(output_text_file_path, 'r', encoding='utf-8') as text_file:
        for line in text_file:
          line = line.encode('latin-1', 'replace').decode('latin-1')
          pdf.cell(200, 10, txt=line, ln=True)

    pdf.output(output_pdf_file_path)
    print(f"PDF conversion complete. Converted PDF saved to {output_pdf_file_path}")

# Example usage
html_file_path = '/content/sec-edgar-filings/MSFT/10-K/94/full-submission_converted.txt'
output_dir = '/content/sec-edgar-filings/MSFT/10-K/94'
convert_html_to_text(html_file_path, output_dir,1994)
html_file_path = '/content/sec-edgar-filings/MSFT/10-K/95/full-submission_converted.txt'
output_dir = '/content/sec-edgar-filings/MSFT/10-K/95'
convert_html_to_text(html_file_path, output_dir,1995)
html_file_path = '/content/sec-edgar-filings/MSFT/10-K/96/full-submission_converted.txt'
output_dir = '/content/sec-edgar-filings/MSFT/10-K/96'
convert_html_to_text(html_file_path, output_dir,1996)
html_file_path = '/content/sec-edgar-filings/MSFT/10-K/97/full-submission_converted.txt'
output_dir = '/content/sec-edgar-filings/MSFT/10-K/97'
convert_html_to_text(html_file_path, output_dir,1997)
html_file_path = '/content/sec-edgar-filings/MSFT/10-K/98/full-submission_converted.txt'
output_dir = '/content/sec-edgar-filings/MSFT/10-K/98'
convert_html_to_text(html_file_path, output_dir,1998)
html_file_path = '/content/sec-edgar-filings/MSFT/10-K/99/full-submission_converted.txt'
output_dir = '/content/sec-edgar-filings/MSFT/10-K/99'
convert_html_to_text(html_file_path, output_dir,1999)
html_file_path = '/content/sec-edgar-filings/MSFT/10-K/00/full-submission_converted.txt'
output_dir = '/content/sec-edgar-filings/MSFT/10-K/00'
convert_html_to_text(html_file_path, output_dir,2000)
html_file_path = '/content/sec-edgar-filings/MSFT/10-K/01/full-submission_converted.txt'
output_dir = '/content/sec-edgar-filings/MSFT/10-K/01'
convert_html_to_text(html_file_path, output_dir,2001)
html_file_path = '/content/sec-edgar-filings/MSFT/10-K/02/full-submission_converted.txt'
output_dir = '/content/sec-edgar-filings/MSFT/10-K/02'
convert_html_to_text(html_file_path, output_dir,2002)
html_file_path = '/content/sec-edgar-filings/MSFT/10-K/03/full-submission_converted.txt'
output_dir = '/content/sec-edgar-filings/MSFT/10-K/03'
convert_html_to_text(html_file_path, output_dir,2003)
html_file_path = '/content/sec-edgar-filings/MSFT/10-K/04/full-submission_converted.txt'
output_dir = '/content/sec-edgar-filings/MSFT/10-K/04'
convert_html_to_text(html_file_path, output_dir,2004)
html_file_path = '/content/sec-edgar-filings/MSFT/10-K/05/full-submission_converted.txt'
output_dir = '/content/sec-edgar-filings/MSFT/10-K/05'
convert_html_to_text(html_file_path, output_dir,2005)
html_file_path = '/content/sec-edgar-filings/MSFT/10-K/06/full-submission_converted.txt'
output_dir = '/content/sec-edgar-filings/MSFT/10-K/06'
convert_html_to_text(html_file_path, output_dir,2006)
html_file_path = '/content/sec-edgar-filings/MSFT/10-K/07/full-submission_converted.txt'
output_dir = '/content/sec-edgar-filings/MSFT/10-K/07'
convert_html_to_text(html_file_path, output_dir,2007)
html_file_path = '/content/sec-edgar-filings/MSFT/10-K/08/full-submission_converted.txt'
output_dir = '/content/sec-edgar-filings/MSFT/10-K/08'
convert_html_to_text(html_file_path, output_dir,2008)
html_file_path = '/content/sec-edgar-filings/MSFT/10-K/09/full-submission_converted.txt'
output_dir = '/content/sec-edgar-filings/MSFT/10-K/09'
convert_html_to_text(html_file_path, output_dir,2009)
html_file_path = '/content/sec-edgar-filings/MSFT/10-K/10/full-submission_converted.txt'
output_dir = '/content/sec-edgar-filings/MSFT/10-K/10'
convert_html_to_text(html_file_path, output_dir,2010)
html_file_path = '/content/sec-edgar-filings/MSFT/10-K/11/full-submission_converted.txt'
output_dir = '/content/sec-edgar-filings/MSFT/10-K/11'
convert_html_to_text(html_file_path, output_dir,2011)
html_file_path = '/content/sec-edgar-filings/MSFT/10-K/12/full-submission_converted.txt'
output_dir = '/content/sec-edgar-filings/MSFT/10-K/12'
convert_html_to_text(html_file_path, output_dir,2012)
html_file_path = '/content/sec-edgar-filings/MSFT/10-K/13/full-submission_converted.txt'
output_dir = '/content/sec-edgar-filings/MSFT/10-K/13'
convert_html_to_text(html_file_path, output_dir,2013)
html_file_path = '/content/sec-edgar-filings/MSFT/10-K/14/full-submission_converted.txt'
output_dir = '/content/sec-edgar-filings/MSFT/10-K/14'
convert_html_to_text(html_file_path, output_dir,2014)
html_file_path = '/content/sec-edgar-filings/MSFT/10-K/15/full-submission_converted.txt'
output_dir = '/content/sec-edgar-filings/MSFT/10-K/15'
convert_html_to_text(html_file_path, output_dir,2015)
html_file_path = '/content/sec-edgar-filings/MSFT/10-K/16/full-submission_converted.txt'
output_dir = '/content/sec-edgar-filings/MSFT/10-K/16'
convert_html_to_text(html_file_path, output_dir,2016)
html_file_path = '/content/sec-edgar-filings/MSFT/10-K/17/full-submission_converted.txt'
output_dir = '/content/sec-edgar-filings/MSFT/10-K/17'
convert_html_to_text(html_file_path, output_dir,2017)
html_file_path = '/content/sec-edgar-filings/MSFT/10-K/18/full-submission_converted.txt'
output_dir = '/content/sec-edgar-filings/MSFT/10-K/18'
convert_html_to_text(html_file_path, output_dir,2018)
html_file_path = '/content/sec-edgar-filings/MSFT/10-K/19/full-submission_converted.txt'
output_dir = '/content/sec-edgar-filings/MSFT/10-K/19'
convert_html_to_text(html_file_path, output_dir,2019)
html_file_path = '/content/sec-edgar-filings/MSFT/10-K/20/full-submission_converted.txt'
output_dir = '/content/sec-edgar-filings/MSFT/10-K/20'
convert_html_to_text(html_file_path, output_dir,2020)
html_file_path = '/content/sec-edgar-filings/MSFT/10-K/21/full-submission_converted.txt'
output_dir = '/content/sec-edgar-filings/MSFT/10-K/21'
convert_html_to_text(html_file_path, output_dir,2021)
html_file_path = '/content/sec-edgar-filings/MSFT/10-K/22/full-submission_converted.txt'
output_dir = '/content/sec-edgar-filings/MSFT/10-K/22'
convert_html_to_text(html_file_path, output_dir,2022)
html_file_path = '/content/sec-edgar-filings/MSFT/10-K/23/full-submission_converted.txt'
output_dir = '/content/sec-edgar-filings/MSFT/10-K/23'
convert_html_to_text(html_file_path, output_dir,2023)

```
